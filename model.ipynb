```python
import sqlite3

# Create a connection to the SQLite database (stored locally in the Colab environment)
conn = sqlite3.connect('atliq_tshirts.db')
cursor = conn.cursor()

# Create the t_shirts table
cursor.execute('''
CREATE TABLE IF NOT EXISTS t_shirts (
    t_shirt_id INTEGER PRIMARY KEY AUTOINCREMENT,
    brand TEXT NOT NULL CHECK (brand IN ('Van Huesen', 'Levi', 'Nike', 'Adidas')),
    color TEXT NOT NULL CHECK (color IN ('Red', 'Blue', 'Black', 'White')),
    size TEXT NOT NULL CHECK (size IN ('XS', 'S', 'M', 'L', 'XL')),
    price INTEGER CHECK (price BETWEEN 10 AND 50),
    stock_quantity INTEGER NOT NULL,
    UNIQUE (brand, color, size)
)
''')

# Create the discounts table
cursor.execute('''
CREATE TABLE IF NOT EXISTS discounts (
    discount_id INTEGER PRIMARY KEY AUTOINCREMENT,
    t_shirt_id INTEGER NOT NULL,
    pct_discount REAL CHECK (pct_discount BETWEEN 0 AND 100),
    FOREIGN KEY (t_shirt_id) REFERENCES t_shirts(t_shirt_id)
)
''')

# Commit changes
conn.commit()

# Function to insert random t-shirt records (simulating stored procedure behavior in SQLite)
import random

brands = ['Van Huesen', 'Levi', 'Nike', 'Adidas']
colors = ['Red', 'Blue', 'Black', 'White']
sizes = ['XS', 'S', 'M', 'L', 'XL']

def populate_tshirts(num_records=100):
    for _ in range(num_records):
        brand = random.choice(brands)
        color = random.choice(colors)
        size = random.choice(sizes)
        price = random.randint(10, 50)
        stock_quantity = random.randint(10, 100)

        # Use try/except to avoid inserting duplicate brand, color, size combinations
        try:
            cursor.execute('''
                INSERT INTO t_shirts (brand, color, size, price, stock_quantity)
                VALUES (?, ?, ?, ?, ?)
            ''', (brand, color, size, price, stock_quantity))
        except sqlite3.IntegrityError:
            # Skip if the record already exists due to the UNIQUE constraint
            pass

    conn.commit()

# Populate t_shirts table with 100 random records
populate_tshirts()

# Insert at least 10 records into the discounts table
cursor.executemany('''
INSERT INTO discounts (t_shirt_id, pct_discount)
VALUES (?, ?)
''', [
    (1, 10.00),
    (2, 15.00),
    (3, 20.00),
    (4, 5.00),
    (5, 25.00),
    (6, 10.00),
    (7, 30.00),
    (8, 35.00),
    (9, 40.00),
    (10, 45.00)
])

# Commit changes
conn.commit()

# Query and display t_shirts and discounts tables
cursor.execute('SELECT * FROM t_shirts')
t_shirts = cursor.fetchall()
cursor.execute('SELECT * FROM discounts')
discounts = cursor.fetchall()

import pandas as pd

# Display the t_shirts table
t_shirts_df = pd.DataFrame(t_shirts, columns=['t_shirt_id', 'brand', 'color', 'size', 'price', 'stock_quantity'])
print("T-Shirts Table:")
print(t_shirts_df)

# Display the discounts table
discounts_df = pd.DataFrame(discounts, columns=['discount_id', 't_shirt_id', 'pct_discount'])
print("\nDiscounts Table:")
print(discounts_df)

# Close the connection to the database
conn.close()
```
